```python
import sqlite3

conn = sqlite3.connect("answer_book.db")
conn.execute("PRAGMA foreign_keys = ON")
cursor = conn.cursor()
print("資料庫連接成功")
cursor.execute("SELECT * FROM AnswerType")

rows = cursor.fetchall()
for row in rows:
    print(row)

cursor.execute("""
SELECT id, weight
FROM AnswerType
""")

types = cursor.fetchall()
```

```python
import random

cursor.execute("""
SELECT id, weight
FROM AnswerType
""")

types = cursor.fetchall()

type_id = random.choices(
    [row[0] for row in types],
    weights=[row[1] for row in types]
)[0]

cursor.execute("""
SELECT id, content
FROM Template
WHERE type_id = ?
ORDER BY RANDOM()
LIMIT 1
""",(type_id,))

template_id, template = cursor.fetchone()

cursor.execute("""
SELECT slot_name, category_id
FROM TemplateSlot
WHERE template_id = ?
""",(template_id,))

slots = cursor.fetchall()

fill_data = {}

for slot_name, category_id in slots:

    cursor.execute("""
    SELECT content
    FROM Word
    WHERE category_id = ?
    ORDER BY RANDOM()
    LIMIT 1
    """,(category_id,))

    fill_data[slot_name] = cursor.fetchone()[0]

result = template.format(**fill_data)

print(result)
```

```python
import sqlite3

def safe_connect():
    conn = sqlite3.connect("answer_book.db")

    try:
        conn.execute("PRAGMA integrity_check;")
    except:
        print("DB 壞了，正在重建...")
        conn.close()

        conn = sqlite3.connect("answer_book.db")
        with open("init.sql", "r", encoding="utf-8") as f:
            conn.executescript(f.read())
        conn.commit()

    return conn
```
